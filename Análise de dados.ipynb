### Tratamento dos dados

A seguradora atende apenas imóveis residenciais, para isso é necessário tratar a base para isolar os demais imóveis

```python
# Apresentando os valores únicos

dados.Tipo.unique()
```

```python
# Removendo duplicados

list(dados.Tipo.drop_duplicates())
```

```python
tipo_de_imovel = dados.Tipo.drop_duplicates()

tipo_de_imovel.index = range(tipo_de_imovel.shape[0])
```

```python
residenciais = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
selecao_residencial = dados['Tipo'].isin(residenciais)
```

```python
dados_residencial = dados[selecao_residencial]
```

```python
dados_residencial.head(15)
```

```python
dados_residencial.Tipo.unique()
```

## Exportando a base de dados apenas com registros residenciais
